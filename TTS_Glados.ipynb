```python
from piper import PiperVoice
import wave # Import the wave module

# Charger le modèle
voice = PiperVoice.load("/content/fr_FR-glados-medium.onnx")

# Texte à convertir en audio
text = "Je vais tous vous tuer !"

# Générer la synthèse vocale
with wave.open("output.wav", "wb") as wav_file: # Open file using wave module
    wav_file.setframerate(voice.config.sample_rate) # Set framerate before synthesize
    wav_file.setsampwidth(2)  # 16-bit
    wav_file.setnchannels(1)  # mono

    # The synthesize method likely writes directly to the file object
    # and doesn't return audio data.
    # Remove this line: audio_data = voice.synthesize(text, wav_file)
    voice.synthesize(text, wav_file)  # Call synthesize to write to file
    # Remove this line: wav_file.writeframes(audio_data) # As audio is already written


print("Synthèse terminée ! Le fichier output.wav est créé.")
```
